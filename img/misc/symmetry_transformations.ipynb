```python
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = 0.3
d = 1
T = 5.0

x = np.arange(-10, 10, 0.1)
y = np.arange(-10, 10, 0.1)
X, Y = np.meshgrid(x, y)
```

```python
fig = plt.figure(1, figsize=(8, 4))

h1 = plt.subplot(2, 4, 1)
Z = np.exp(-a * ((X + d) ** 2 + Y**2)) + np.exp(-((X - d) ** 2 + Y**2))
plt.contour(X, Y, Z, linewidths=0.25)
plt.arrow(0, 0, 2.5, 0, head_width=0.5, head_length=0.5, fc="k", ec="k")
plt.arrow(0, 0, 0, 2.5, head_width=0.5, head_length=0.5, fc="k", ec="k")
plt.setp(
    h1,
    xlim=(-9, 9),
    ylim=(-8, 8),
    xticks=([-5, 0, 5]),
    xticklabels=[],
    yticks=([-5, 0, 5]),
    yticklabels=[r"$-a$", r"$0$", r"$a$"],
)

h2 = plt.subplot(2, 4, 2)
Z = np.exp(-a * ((X + d - T) ** 2 + Y**2)) + np.exp(-((X - d - T) ** 2 + Y**2))
plt.contour(X, Y, Z, linewidths=0.25)
plt.arrow(0, 0, 2.5, 0, head_width=0.5, head_length=0.5, fc="k", ec="k")
plt.arrow(0, 0, 0, 2.5, head_width=0.5, head_length=0.5, fc="k", ec="k")
plt.setp(
    h2,
    xlim=(-9, 9),
    ylim=(-8, 8),
    xticks=([-5, 0, 5]),
    xticklabels=[],
    yticks=([-5, 0, 5]),
    yticklabels=[],
)

h3 = plt.subplot(2, 4, 3)
Z = np.exp(-a * ((Y + d - T) ** 2 + X**2)) + np.exp(-((Y - d - T) ** 2 + X**2))
plt.contour(X, Y, Z, linewidths=0.25)
plt.arrow(0, 0, 2.5, 0, head_width=0.5, head_length=0.5, fc="k", ec="k")
plt.arrow(0, 0, 0, 2.5, head_width=0.5, head_length=0.5, fc="k", ec="k")
plt.setp(
    h3,
    xlim=(-9, 9),
    ylim=(-8, 8),
    xticks=([-5, 0, 5]),
    xticklabels=[],
    yticks=([-5, 0, 5]),
    yticklabels=[],
)

h4 = plt.subplot(2, 4, 4)
Z = np.exp(-a * ((-Y + d - T) ** 2 + (-X) ** 2)) + np.exp(
    -((-Y - d - T) ** 2 + (-X) ** 2)
)
plt.contour(X, Y, Z, linewidths=0.25)
plt.arrow(0, 0, 2.5, 0, head_width=0.5, head_length=0.5, fc="k", ec="k")
plt.arrow(0, 0, 0, 2.5, head_width=0.5, head_length=0.5, fc="k", ec="k")
plt.setp(
    h4,
    xlim=(-9, 9),
    ylim=(-8, 8),
    xticks=([-5, 0, 5]),
    xticklabels=[],
    yticks=([-5, 0, 5]),
    yticklabels=[],
)

h5 = plt.subplot(2, 4, 5)
Z = np.exp(-a * ((X + d) ** 2 + Y**2)) + np.exp(-((X - d) ** 2 + Y**2))
plt.contour(X, Y, Z, linewidths=0.25)
plt.arrow(0, 0, 2.5, 0, head_width=0.5, head_length=0.5, fc="k", ec="k")
plt.arrow(0, 0, 0, 2.5, head_width=0.5, head_length=0.5, fc="k", ec="k")
plt.setp(
    h5,
    xlim=(-9, 9),
    ylim=(-8, 8),
    xticks=([-5, 0, 5]),
    xticklabels=[r"$-a$", r"$0$", r"$a$"],
    yticks=([-5, 0, 5]),
    yticklabels=[r"$-a$", r"$0$", r"$a$"],
)

h6 = plt.subplot(2, 4, 6)
Z = np.exp(-a * ((X + d) ** 2 + Y**2)) + np.exp(-((X - d) ** 2 + Y**2))
plt.contour(X, Y, Z, linewidths=0.25)
plt.arrow(-5, 0, 2.5, 0, head_width=0.5, head_length=0.5, fc="k", ec="k")
plt.arrow(-5, 0, 0, 2.5, head_width=0.5, head_length=0.5, fc="k", ec="k")
# text(-1,-1.5,r'$x$')
# text(-6.5,4,r'$y$')
plt.setp(
    h6,
    xlim=(-9, 9),
    ylim=(-8, 8),
    xticks=([-5, 0, 5]),
    xticklabels=[r"$-a$", r"$0$", r"$a$"],
    yticks=([-5, 0, 5]),
    yticklabels=[],
)

h7 = plt.subplot(2, 4, 7)
Z = np.exp(-a * ((X + d) ** 2 + Y**2)) + np.exp(-((X - d) ** 2 + Y**2))
plt.contour(X, Y, Z, linewidths=0.25)
plt.arrow(-5, 0, 2.5, 0, head_width=0.5, head_length=0.5, fc="k", ec="k")
plt.arrow(-5, 0, 0, -2.5, head_width=0.5, head_length=0.5, fc="k", ec="k")
plt.setp(
    h7,
    xlim=(-9, 9),
    ylim=(-8, 8),
    xticks=([-5, 0, 5]),
    xticklabels=[r"$-a$", r"$0$", r"$a$"],
    yticks=([-5, 0, 5]),
    yticklabels=[],
)

h8 = plt.subplot(2, 4, 8)
Z = np.exp(-a * ((X + d) ** 2 + Y**2)) + np.exp(-((X - d) ** 2 + Y**2))
plt.contour(X, Y, Z, linewidths=0.25)
plt.arrow(-5, 0, -2.5, 0, head_width=0.5, head_length=0.5, fc="k", ec="k")
plt.arrow(-5, 0, 0, 2.5, head_width=0.5, head_length=0.5, fc="k", ec="k")
plt.setp(
    h8,
    xlim=(-9, 9),
    ylim=(-8, 8),
    xticks=([-5, 0, 5]),
    xticklabels=[r"$-a$", r"$0$", r"$a$"],
    yticks=([-5, 0, 5]),
    yticklabels=[],
)

plt.subplots_adjust(wspace=0, hspace=0)

plt.figtext(0.18, 0.92, "Original")
plt.figtext(0.36, 0.92, "Translation")
plt.figtext(0.57, 0.92, "Rotation")
plt.figtext(0.76, 0.92, "Inversion")

plt.figtext(0.07, 0.65, "Active", rotation=90)
plt.figtext(0.07, 0.25, "Passive", rotation=90)

plt.savefig("symmetry_transformations.svg")
```
