```python
import veloxchem as vlx
```

```python
molstr = """O   0.0   0.0   0.0
H   0.0   1.4   1.1
H   0.0  -1.4   1.1"""
molecule = vlx.Molecule.read_molecule_string(molstr, units='au')
basis = vlx.MolecularBasis.read(molecule, 'def2-svpd')
```

```python
scf_drv = vlx.ScfRestrictedDriver()
scf_drv.xcfun = 'pbe0'
scf_results = scf_drv.compute(molecule, basis)
```

```python
rpa_drv = vlx.LinearResponseEigenSolver()
rpa_drv.nstates = 3
rpa_drv.nto = True
rpa_results = rpa_drv.compute(molecule, basis, scf_results)
```

```python
print(rpa_results['nto_h5_files'])
```

```python
nto_mo = vlx.MolecularOrbitals.read_hdf5(rpa_results['nto_h5_files'][0])
orbviewer = vlx.OrbitalViewer()
orbviewer.plot(molecule, basis, nto_mo)
```
